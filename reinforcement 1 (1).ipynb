 Data cleaning

```python
data.isnull().sum()
```

```python
# calculate the no of null values in dataset
data.isnull().sum().sum()
```

```python
#define columns fill with median values
median_cols=['sqft_living','sqft_lot','yr_built']
for col in median_cols:
    data[col].fillna(data[col].median(),inplace=True)
    
```

```python
#fill city column with the most frequent values
data['city'].fillna(data['city'].mode()[0],inplace=True)
print(data.isnull().sum())
```

```python
data.drop_duplicates()
```

```python
#drop the time ,no values to analyses
data['date'] = pd.to_datetime(data['date'],errors='coerce')
data.head()
# drop the country columns 
data=data.drop('country',axis=1)
data.head()


```

```python
#change the column to (int)
columns_to_convert=['bathrooms','floors','price','waterfront','yr_built']
for col in columns_to_convert:
    data[col]=data[col].astype(int)
data.head()

```

```python
data.dtypes
```

```python
view_mapping = {
    0: "No view",
    1: "Poor view",
    2: "Average view",
    3: "Good view",
    4: "Excellent view"
}
data['view']=data['view'].map(view_mapping).astype(str)
```

```python
condition_mapping = {
    1: "Poor",
    2: "Fair",
    3: "Average",
    4: "Good",
    5: "Excellent"
}

# Apply the mapping to the 'condition' column
data['condition'] = data['condition'].map(condition_mapping).astype(str)
```

```python
waterfront={0:"No",1:"Yes"}
data['waterfront'] = data['waterfront'].map(waterfront).astype(str)
data.head()
```

```python
# check for outliers
import matplotlib.pyplot as plt
# select the columns
colums=['price','bedrooms','bathrooms','sqft_living','sqft_lot','floors','sqft_above','sqft_basement','yr_built','yr_renovated']
#create Boxplot
plt.figure(figsize=(16,6))
data[colums].boxplot(rot=50)
plt.grid(False)
plt.ylabel('values')
plt.title('box plot for numerical columns')
plt.show()
```

```python
numeric_col=data.select_dtypes(include=['number'])
for i in range(10):
    for col in numeric_col.columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        
        lower_bound = Q1 - 1.5* IQR
        upper_bound = Q3 + 1.5* IQR
        
        print(f"Lower bound:{lower_bound}")
        print(f"Upper bound:{upper_bound}")
        #calculate the median
        median_value = data[col].median()
        print(f"median value:{median_value}")
        #replace outliers
        data.loc[(data[col]<=lower_bound) | (data[col]>=upper_bound),col]=round(median_value,0)
        print("after replacing outliers:",data)
```

```python
colums=['price','bedrooms','bathrooms','sqft_living','sqft_lot','floors','sqft_above','sqft_basement','yr_built','yr_renovated']
#create Boxplot
plt.figure(figsize=(10,3))
data[colums].boxplot(rot=50)
plt.grid(False)
plt.ylabel('values')
plt.title('box plot for numerical columns')
plt.show()
```

```python
#Histogram
columns = ['price', 'floors', 'sqft_living', 'sqft_lot', 'bedrooms', 'bathrooms']
plt.figure(figsize=(12,10))

# Plot each column's distribution
for col in columns:
    plt.figure(figsize=(5, 4))
    data[col].hist(bins=30, color='lightgreen', edgecolor='green')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

 Feature Engineering:

```python
#  average price per city
print("Average price of city:")
data.groupby('city')['price'].mean()

```

```python
avg_price_city = data.groupby('city')['price'].mean().sort_values(ascending=False).head(5)

avg_price_city.plot(kind='bar', figsize=(15, 5), color='orange')
plt.title('Top 5 Cities by Average Price')
plt.xlabel('City')
plt.ylabel('Average Price')
plt.xticks(rotation=10)
plt.show()
```

```python
from scipy import stats
# 1. ANOVA: Price across Bedroom Counts

grouped_prices = [group['price'].values for name, group in data.groupby('bedrooms')]
f_stat, p_val_anova = stats.f_oneway(*grouped_prices)

print("ANOVA: Price by Number of Bedrooms")
print(f"F-statistic = {f_stat:.3f}, p-value = {p_val_anova:.5e}\n")
```

INSIGHTS



1. Data Exploration and Visualization insights:
This initial step involves using various techniques to understand the dataset.
Data exploration can be done through visual representations like histograms, scatter plots, and box plots to understand the distribution of data, identify outliers, and reveal potential relationships between variables.
For example, one might visualize the relationship between square footage and price or compare the average prices of properties in different locations. 


2. Statistical Analysis:
Statistical methods like regression analysis can be used to quantify the relationship between housing prices and various factors.
For instance, a regression model can estimate the impact of square footage, number of bedrooms, or proximity to amenities on the price of a house. 
This helps determine the relative importance of each factor in determining price. 


3. Key Drivers:
Location: Proximity to desirable amenities like schools, parks, and public transportation, as well as neighborhood characteristics, significantly impact prices. 
Property Size: Larger properties generally command higher prices, but the relationship may not be linear. 
Condition: Properties in good condition, with recent renovations or upgrades, tend to sell for more. 
Amenities: Features like swimming pools, updated kitchens, or modern bathrooms can positively influence price. 


4. Recommendations:
Pricing: Understanding the impact of different factors allows for more accurate pricing strategies. A seller can use the analysis to determine a fair market price or potentially price a property higher if it has desirable features. 
Investment: Investors can use the analysis to identify areas with high growth potential or properties that are undervalued. 
Property Development: Developers can use the analysis to identify areas where there is a demand for specific types of properties or amenities, and to understand the potential return on investment for new developments. 


5. Further Analysis:
Beyond basic factors, demographics, economic conditions (interest rates, income levels), and market trends (supply and demand) also play a role in housing prices.
